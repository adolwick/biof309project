## Extra Visualizations: Year by Year Visualizations

I wanted to separate the scatterplot I made of yearly pace and distance so that the scatterplots for each individual year could be seen. In these extra visualizations, I simply repurposed the code I used to create the main graph.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from dolwick_project import splitdate
from dolwick_project import test_splitdate
```

```python
e = pd.read_csv("clean.csv")
```

```python
eymd = splitdate.splitdate(e, e['Date'])
pd.options.display.max_rows=20
print(eymd)
#making sure splitdate works properly
test_splitdate.tsd()
```

```python
# col_dict = {'2009':'#000000', '2010':'#450842', '2011':'#1B0951', '2012':'#1620CD' , 
            #'2013':'#3AA6FF' , '2014':'#3AFFDE' , '2015':'#3AFF51' , '2016':'#D0FF00',
            #'2017':'#FFC600', '2018':'#FF8600', '2019':'#FF3600'}

fig2009 = plt.figure(figsize=(12,10),dpi=75)
plt.xticks(np.arange(0, max(e['Mileage'])+1, 1.0))
plt.xlim(0,19)
plt.ylim(3.5, 12)
ax2009 = fig2009.gca()
ax2009.scatter(eymd['Mileage'][(eymd['year'] == '2009')],eymd['pace_min'][(eymd['year'] == '2009')],c='#000000',alpha=0.3)
ax2009.set_xlabel('Distance (mi)', fontsize=13)
ax2009.set_ylabel('Pace (min/mi)', fontsize=13)
ax2009.set_title('Pace and Distance of Each Run in 2009', fontsize=20)
```
