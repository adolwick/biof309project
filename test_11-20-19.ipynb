```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("~/Documents/NIH IRTA/Running Log.csv")
```

```python
print(df)
```

```python
df["yada"] = pd.np.where(df.Time.str.contains('\d:\d\d:\d\d', regex = True), df["Time"], "0:" + df["Time"])
print(df)
```

```python
#df.dropna(inplace=True)
new = df["yada"].str.split(pat = ":", expand = True)
print(new)
```

```python
result = pd.concat([df, new], axis=1, sort=False)
print(result)
```

```python
df2 = result.rename(columns={"yada": "time_corr", 0: "hours", 1: "minutes", 2: "seconds"})
print(df2)
```

```python
df2['hours'] = pd.to_numeric(df2['hours'])
df2['minutes'] = pd.to_numeric(df2['minutes'])
df2['seconds'] = pd.to_numeric(df2['seconds'])
```

```python
time_seconds = df2.hours*3600 + df2.minutes*60 + df2.seconds
df2['time_s'] = time_seconds
print(df2)
```

```python
df2['Mileage'] = pd.to_numeric(df2['Mileage'])
pace_s = df2.time_s / df2.Mileage
df2['pace_s'] = pace_s
print(df2.tail(10))
```

```python
pace_min = df2.pace_s / 60
df2['pace_min'] = pace_min
print(df2.tail(10))
```

```python
df2.pace_min = df2.pace_min.astype(str)
```

```python
splits = df2["pace_min"].str.split(pat = ".", expand = True)
print(splits.tail(10))
```
